```python
import sqlalchemy
import noteql


```

```python
session = noteql.Session('newexample')
```

```python
import time
start = time.time()
session.load_json('full-1.1.json', path_to_list='releases', table_name='import', field_name='releases')
end = time.time()
print(end - start)
```

```python
session.run_sql('''select count(*) from import limit 1''')
```

```python
session.run_sql('''
drop table if exists releases;
select  releases ->> 'ocid' as ocid, releases ->> 'id' as release_id, releases as release into releases from import;


drop table if exists awards;
select ocid, release_id, award->>'id' as award_id, award into awards from (
  select ocid, release_id, jsonb_array_elements(release->'awards') as award from releases
) a;

drop table if exists contracts;
select ocid, release_id, contract->>'id' as contract_id, contract into contracts from (
  select ocid, release_id, jsonb_array_elements(release->'contracts') as contract from releases
) a;

drop table if exists buyer;
select ocid, release_id, release->'buyer'->'id' as buyer_id, release->'buyer' as buyer into buyer from releases;

drop table if exists parties;
select ocid, release_id, party->>'id' as organization_id, party into parties from (
  select ocid, release_id, jsonb_array_elements(release->'parties') as party from releases
) a;

drop table if exists tender;
select ocid, release_id, release->'tender'->'id' as tender_id, release->'tender' as tender into tender from releases;

drop table if exists planning;
select ocid, release_id, release->'planning' as planning into planning from releases;

drop table if exists related_processes;
select ocid, release_id, related_process->>'id' as related_process_id, related_process into related_processes from (
  select ocid, release_id, jsonb_array_elements(release->'relatedProcesses') as related_process from releases
) a;
''')
```

```python
session.run_sql('select * from releases limit 1')
```
